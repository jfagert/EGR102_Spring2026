## Matrix Operations Using Python

Now you will work on examples of matrices multiplying vectors to get yourselves comfortable with matrix operations in Python.  First, let's define the matrix $\mathbf{A}$ using Python as follows:

````python
from IPython.display import display
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sympy import Matrix, init_printing
init_printing()

A=np.array([[2,1],[3,-1],[0,4]])
print("A = ")
display(Matrix(A))
````

Note that in Python, we are using a library called "NumPy" to create matrices. At the very beginning of this code, the NumPy library was called using the command:
```python
import numpy as np
```
More information on NumPy can be found at https://numpy.org/.

For the matrix $\mathbf{A}$ above, we are using the "array" function to define each element in the matrix. The general approach is to have each row of values in one set of square brackets [], and each column separated by commas within those square brackets. To select an element or individual row/column from the matrix, we can use indexing and "slicing". To select an individual element, you can index it using a row,column notation. In Python, the first element is at index 0. As an example, the first row, first column, and third row, second column of A can be extracted in the following way:

````python
print(A[0,0])
print(A[2,1])
````

To "slice" the matrix, you can use a similar notation, but call a range of values across rows and/or columns. The notation for slicing is start:stop:step. For example, you may want to extract only the first two rows and the first column of $\mathbf{A}$; this is done in the following way:

````python
print(A[0:2,0])
````

You can also use slicing to select a submatrix. For example, to select/isolate the 2x2 matrix that contains the first two rows and both columns of $\mathbf{A}$, the following commands can be executed. Note that, in this case, all of the columns of $\mathbf{A}$ can be selected by using a colon symbol $(:)$.

````python
print(A[0:2,:])
````

Similar to matrix $\mathbf{A}$ above, we can define both column vectors and row vectors using NumPy. The following code shows how to define a column vector, $\mathbf{v}$ and a row vector, $\mathbf{u}$:

````python
v = np.array([[-2],[1]])
u = np.array([[2,-3,1]])

print("\n v= :")
display(Matrix(v))
print("\n u= :")
display(Matrix(u))
````

### Try it yourself!

Using the definitions for $\mathbf{A}$, $\mathbf{v}$, and $\mathbf{u}$ from above, please solve the following using Python. Do the answers match what you expect? (Not all of these may be defined!)

1. $\mathbf{A}$$\mathbf{v}$
2. $\mathbf{u}$$\mathbf{A}$
3. $\mathbf{A}$$\mathbf{u}$
4. $\mathbf{v}$$\mathbf{A}$
5. (First two rows of $\mathbf{A}$) * $\mathbf{v}$
6. $\mathbf{u}$ * (second column of $\mathbf{A}$)
7. (Second through 4th column of $\mathbf{A}$) * $\mathbf{v}$
8. $\mathbf{u}$* (the first row of $\mathbf{A}$)

````python
# A*v
print("\n 1) A*v = :")
display(Matrix(np.matmul(A,v)))

# u*A
print("\n 2) u*A = :")
display(Matrix(np.matmul(u,A)))

# A*u
try:
    display(Matrix(np.matmul(A,u)))
except Exception:
    print("3) A*u Multiplication is not possible")

# v*A
try:
    display(Matrix(np.matmul(v,A)))
except Exception:
    print("4) v*A Multiplication is not possible")

# First two rows of A * v
print("\n 5) A[0:1]*v = :")
D = np.matmul(A[0:2,:],v)
display(Matrix(D))

# u* second column of A
print("\n 6) u*A[:1] = :")
display(Matrix(np.matmul(u,A[:,1])))

# second through 4th column of A * v
try:
    display(Matrix(np.matmul(A[:,1:3],v)))
except Exception:
    print("7) Indexing exceeds dimensions of A")

# u* the first row of A
try:
    display(Matrix(np.matmul(u,A[0,:])))
except Exception:
    print("8) u*A[0,:] Multiplication is not possible")
````

````python

# Question 1: load csv files
df = pd.read_csv('TempData_Combined.csv',header = None)
df2 = pd.read_csv('MonthYear_1994_2024.csv', header=None)

#convert loaded data to matrices
TempData = df.to_numpy()
MonthData = df2.to_numpy()
monthyear = np.array(MonthData, dtype=str)

#Question 2: determine the size of the temperature data matrix
print(TempData.shape)
````

````python
#Question 3: separate each city's data
t1=TempData[:,0]
print(t1.shape)
t2=TempData[:,1]
print(t2.shape)
t3=TempData[:,2]
print(t3.shape)
````

````python
#Question 5: compute the min and max for each city

print("City 1 Min:",np.min(t1))
print("City 2 Min:",np.min(t2))
print("City 3 Min:",np.min(t3))

print("City 1 Max:",np.max(t1))
print("City 2 Max:",np.max(t2))
print("City 3 Max:",np.max(t3))
````

````python
#Question 6: plot the data
xdata=np.arange(1, (1 + len(TempData))) #create a vector of numbers starting at 1 and going up by one until the length of the temperature data matrix
plt.plot(xdata, TempData[:,0], linestyle='-', color='blue',label='Cleveland')
plt.plot(xdata, TempData[:,1], linestyle='-', color='red',label='NYC')
plt.plot(xdata, TempData[:,2], linestyle='-', color='green',label='San Diego')

# make xticks and xtick labels
xticks = np.arange(0, len(TempData[:,0]), 12) #this makes it so that the xticks are monthly
subset = [monthyear[i] for i in xticks] #uses the names in monthyear to create the xtick labels

new_labels = [subset[0] if isinstance(subset, list) else subset.replace('[', '').replace(']', '') for subset in subset] # this formats the labels to remove the square brackets and quotes - make it look nicer


# Add titles and labels
plt.title("Monthly Temperature Data")
plt.xlabel("Month and Year")
plt.ylabel("Average Temperature, \u00b0F")
plt.legend(loc="lower right")
plt.grid(color='green', linestyle='--', linewidth=0.5, axis='y')
plt.xticks(ticks=xticks,labels=new_labels, rotation=45,ha='right') #rotating the xticks by 45 degrees and aligning them to the right to make them more easily read

# Display the plot
plt.show()
````

````python
#Question 7: temperature prediction

preditFormula=np.array([[-0.316],[0.594]])

SanDiegoPredict=np.matmul(TempData[:,0:2],preditFormula)+45.97

#Question 8: plotting prediction vs. actual curve

xdata=np.arange(1, (1 + len(TempData))) #create a vector of numbers starting at 1 and going up by one until the length of the temperature data matrix
plt.plot(xdata, SanDiegoPredict, linestyle='-', color='blue',label='San Diego Prediction')
plt.plot(xdata, TempData[:,2], linestyle='-', color='green',label='San Diego')

# make xticks and xtick labels
xticks = np.arange(0, len(TempData[:,0]), 12) #this makes it so that the xticks are monthly
subset = [monthyear[i] for i in xticks] #uses the names in monthyear to create the xtick labels

new_labels = [subset[0] if isinstance(subset, list) else subset.replace('[', '').replace(']', '') for subset in subset] # this formats the labels to remove the square brackets and quotes - make it look nicer


# Add titles and labels
plt.title("Monthly Temperature Data")
plt.xlabel("Month and Year")
plt.ylabel("Average Temperature, \u00b0F")
plt.legend(loc="lower right")
plt.grid(color='green', linestyle='--', linewidth=0.5, axis='y')
plt.xticks(ticks=xticks,labels=new_labels, rotation=45,ha='right') #rotating the xticks by 45 degrees and aligning them to the right to make them more easily read

# Display the plot
plt.show()
````
