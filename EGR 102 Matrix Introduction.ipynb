```python
import matplotlib.pyplot as plt
import numpy as np
# from sympy import *
# import pprint as pp

# Define vector components [u, v]
vectorV = np.array([[1],[0]])
# vectorv=Matrix([[1],[0]])

# pp.pprint(vectorV)
print("V=",vectorV)

# Plot the vector starting from the origin (0, 0)
plt.quiver(0, 0, vectorV[0], vectorV[1], angles='xy', scale_units='xy', scale=1, color='blue')

# Set axis limits for better viewing
plt.xlim(0, 3)
plt.ylim(0, 3)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("2D Vector Visualization")
plt.grid()
plt.show()
```

Next, transform vector v with matrix A

```python
vectorA = np.array([[1/(2**.5), -1/(2**.5)],[1/(2**.5), 1/(2**.5)]])
print("A=",vectorA)

vectorW=np.matmul(vectorA,vectorV)
print("W=",vectorW)

plt.quiver(0, 0, vectorW[0], vectorW[1], angles='xy', scale_units='xy', scale=1, color='blue')

# Set axis limits for better viewing
plt.xlim(0, 3)
plt.ylim(0, 3)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("2D Vector Visualization")
plt.grid()
plt.show()
```
